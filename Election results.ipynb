```python
%matplotlib inline
import pandas as pd
import seaborn as sns
from io import BytesIO
```

```python
sample_results_filepath = "Raw data/csv files/Izbori 2010/PARLAMENT BiH/FBiH/2010_FBiH_511_IZBORNA_JEDINICA_1.csv"
```

```python
# Alternatively BytesIO can be used: http://stackoverflow.com/questions/13690122/pandas-read-csv-and-utf-16

def load_election_result(file):
    frame = pd.read_csv(file, sep='\t', index_col=0, encoding='utf-16', header=0)
    # Drop empty rows http://thomas-cokelaer.info/blog/2014/05/pandas-read_csv-how-to-skip-empty-lines/
    frame.dropna(how='all', inplace=True)
    return frame
    
frame = load_election_result(sample_results_filepath)
frame.head(2)
```
